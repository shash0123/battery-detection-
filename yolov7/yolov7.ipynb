```python
import zipfile
import os

# Path to your downloaded zip file
zip_path = '/content/battery detection  2.v3i.yolov7pytorch.zip'  # Replace with the actual path

# Destination path to extract the files
extract_dir = '/content/battery-detection/'  # You can adjust the directory

# Unzip the dataset
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print("Dataset unzipped successfully!")
```

```python
import shutil
import os

# List of source files to move and rename
source_files = ['/content/battery-detection/train/images', '/content/battery-detection/valid/images']  # Replace with your file paths

# List of movie names to rename the files to
movie_titles = ['train', 'valid',]  # Replace with your movie names

# Destination folder
destination_folder = '/content/images/'

# Loop through the files and rename them after movies
for i, file in enumerate(source_files):
    # Create new file name using movie title
    new_name = movie_titles[i] + os.path.splitext(file)[1]  # Retains the file extension
    new_path = os.path.join(destination_folder, new_name)

    # Move and rename the file
    shutil.move(file, new_path)

# Verify if the files have been renamed and moved
os.listdir(destination_folder)
```
